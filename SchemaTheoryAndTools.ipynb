# Building models
1. Specifying models: model notation!
- list variables, give type and range
- label as measured and latent
- label as predictors or predicted
- use behavior theory to connect variables

Example:
y:  response, {0,1}, measured output 
z:  responsevalue, (-inf, inf), latent
c:  context, [subjid, conditionid, stimid], measured
o:  outcomevalues[1:m], outcomeid=1:k -> response = 0, outcomeid = k+1:m -> response = 1, value[1:m]), latent value, types modeled
x:  stimulus, vector[n]   measured input

y ~ glm(z)
z[c] ~ sum( o[c] )
o[c] = sum( x ) 

To use linear model like above, transform measured variables
